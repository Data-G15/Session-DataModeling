### Introducci칩n

En la pr치ctica, es raro que un an치lisis de datos involucre solamente una tabla. Normalmente tendr치s muchas tablas contribuyendo a tu an치lisis y necesitas herramientas flexibles para combinarlas.

Si ha utilizado alg칰n lenguaje de consulta a una base de datos, como SQL, recordar치 que hay ciertas operaciones que permiten unificar dos o m치s de esas tablas, complementando los valores entre ellas. A este tipo de operaci칩n se le llama **JOIN**

Hay 3 operaciones fundamentales que se pueden realizar con un **JOIN**:

- **Uniones uno a uno:** por ejemplo, al unir dos tablas por su columna de 칤ndices o ID (que debe contener valores 칰nicos).

- **Uniones de muchos a uno:** por ejemplo, cuando se une un 칤ndice (칰nico) a una o m치s columnas en una tabla diferente.

- **Uniones de muchos a muchos:** uni칩n donde el criterio de 칤ndices abarca dos o m치s columnas. Por ejemplo, cuando queremos unir una tabla de aviones con una de vuelos y necesitamos unir mediante la aerol칤nea y el ID de vuelo, pues si solo lo hacemos mediante una de ellas, podr칤a haber duplicados o no haber coincidencias.

#### Nota:
Al unir columnas en columnas (potencialmente una combinaci칩n de muchos a muchos), se descartar치n los 칤ndices de las tablas pasadas.

Vale la pena dedicar un tiempo a comprender el resultado del caso de uni칩n de muchos a muchos. ## Merge methods

El argumento `how` del m칠todo `merge` especifica c칩mo determinar qu칠 variables (columnas) ser치n incluidas en la tabla resultante. Si una valores no aparece en las tablas de la izquierda o de la derecha, los valores de la tabla combinada ser치n `NA`. Aqu칤 hay un resumen de las opciones de `how` y sus nombres equivalentes de SQL:

| Merge method | Nombre del SQL Join | Descripci칩n                                  |
| ------------ | ------------------- | -------------------------------------------- |
| left         | LEFT OUTER JOIN     | Usa 칰nicamente las variables de la izquierda |
| right        | RIGHT OUTER JOIN    | Usa 칰nicamente las variables de la derecha   |
| outer        | FULL OUTER JOIN     | Utiliza la uni칩n de las variable de ambos data frames |
| inner        | INNER JOIN          | Utiliza la intersecci칩n de las variables de ambos data frames |

### Left Join

```python
pd.merge(left2, right2 , how="left", on = ['key1', 'key2'])
```

### Right Join

```python
pd.merge(left2, right2 , how="right", on = ['key1', 'key2'])
```

### Outer Join

```python
pd.merge(left2, right2, how="outer", on=["key1", "key2"])
```

### Inner Join

```python
pd.merge(left2, right2, how="inner", on=["key1", "key2"])
```

#### Nota:



Esos fueron los tipos de join que puedes realizar entre data frames.

El m칠todo `merge` acepta m치s argumentos en funci칩n de los casos especiales que se te vayan presentando. Para esos casos, te recomiendo consultar la [documentaci칩n](https://pandas.pydata.org/docs/user_guide/merging.html#database-style-dataframe-or-named-series-joining-merging).
